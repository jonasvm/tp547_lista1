3) Um fabricante de pistões de metal descobre que, em média, 15% de seus pistões são rejeitados porque são superdimensionados ou subdimensionados. Qual é a probabilidade de um lote de 8 pistões conter

(a) não mais que 2 rejeitados?

```python
import numpy as np

value=2
count=0
q=0.15    #Probabilidade de acerto
n=8      #Número de tentativas
N=100000 #Número de amostras
c = q/(1-q)
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = pow((1 - q),n)
    F = pr
    while ix>=F:
        pr = (c * (n - i) / (i + 1))* pr;
        F = F + pr;
        i = i + 1;
    a1=i
    av=np.append(av,a1)
    if i<= value: count+=1

#print(av)
print(count/N)
```

(b) pelo menos 6 rejeitados?

```python
import numpy as np

value=6
count=0
q=0.15    #Probabilidade de acerto
n=8      #Número de tentativas
N=100000 #Número de amostras
c = q/(1-q)
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = pow((1 - q),n)
    F = pr
    while ix>=F:
        pr = (c * (n - i) / (i + 1))* pr;
        F = F + pr;
        i = i + 1;
    a1=i
    av=np.append(av,a1)
    if i >= value: count+=1

#print(av)
print(count/N)
```

- Traçar o histograma da variável analisada.

```python
import matplotlib.pyplot as plt

# Criar o histograma
plt.hist(av, bins=range(int(av.min()), int(av.max()) + 1), color='r')

# Adicionar os rótulos(labels)
plt.title('Histograma dos Pistões Rejeitados')
plt.xlabel('Rejeitados')
plt.ylabel('Densidade')

# Exibir o histograma
plt.show()
```
