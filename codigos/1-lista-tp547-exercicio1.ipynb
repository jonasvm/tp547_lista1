1. Mostre, usando análise e simulação, que o gerador de números aleatórios definido por Xi+1 = 5xi mod (7) é um gerador de período completo. Determine a sequência gerada para sementes x0 = 4 e x0 = 7.
Compare as sequências e comente os resultados.

Para x0=4

```python
import numpy as np
import matplotlib.pyplot as plt

x=4
x1=np.array([x])
n=50
a=5
m=7
for i in range(n):
    x=(a*x)%m
    x1=np.append(x1,x)
print(x1)
ind=np.arange(n+1)
plt.bar(ind, x1)
plt.show()
```

Para x0=7

```python
import numpy as np
import matplotlib.pyplot as plt

x=7
x1=np.array([x])
n=50
a=5
m=7
for i in range(n):
    x=(a*x)%m
    x1=np.append(x1,x)
print(x1)
ind=np.arange(n+1)
plt.bar(ind, x1)
plt.show()
```
