2. O número de chamadas para o help-desk de uma empresa tem uma distribuição de Poisson com 60 chamadas por um período de 10 horas. Se C = a variável aleatória para o número de chamadas por hora, encontre:

a. A probabilidade de que o suporte técnico não receba chamadas em uma determinada hora.

```python
import numpy as np

lambda1=6 #Número médio de requisições (60 chamadas em 10 horas = 6 chamadas por hora em média)
N=100000 #Numero de amostras
value=0
count=0
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix>=F:
        pr=lambda1/(i+1)*pr
        F = F + pr
        i = i + 1;
    a1=i
    av=np.append(av,a1)
    if i == value: count += 1

print(count/N)
```
