4) Se ocorrerem falhas de energia elétrica de acordo com uma distribuição de Poisson com uma média de 6 falhas a cada duas semanas, calcule a probabilidade de que haverá ao menos 2 falhas durante uma semana específica.

```python
import numpy as np



lambda1=3 #Número médio de requisições
N=100000  #Numero de amostras
value=2
count=0

av=np.array([])
x=np.random.uniform(0,1,N)

for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix>=F:
        pr=lambda1/(i+1)*pr
        F = F + pr
        i = i + 1;
    a1=i
    av=np.append(av,a1)
    if i>=value : count += 1

#print(av)
print(count/N)
```

Traçar o histograma da variável analisada.

```python
import matplotlib.pyplot as plt

# Criar o histograma
plt.hist(av, bins=range(int(av.min()), int(av.max()) + 1), color='g')

# Adicionar os rótulos(labels)
plt.title('Histograma das Falhas de Energia')
plt.xlabel('Falhas')
plt.ylabel('Densidade')

# Exibir o histograma
plt.show()
```
