```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000                                #meu espaço amostral
values = np.random.uniform(0, 1, n)       #gerando os valores
fx_inv = np.log((np.exp(2) - 1)*values + 1)   #minha função inversa

X=np.arange(0, 2, 0.1)                    #gerando valores de X de 0 a 2 com um passo de 0.1
fx = np.exp(X)/(np.exp(2)-1)              #calculando os valores de fx, de acordo com a função de Fx
plt.plot(X, fx)                           #plotando os valores

# Criando o gráfico
plt.hist(fx_inv, bins=50, color='g')

# Exibir o grafico
plt.show()
```
